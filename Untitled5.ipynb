```python
import pandas as pd
df = pd.read_csv('splitwise_benchmark.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(df['n'], df['NumTransactions_Greedy'], 'o-', label='Greedy', linewidth=2)
plt.plot(df['n'], df['NumTransactions_Backtracking'], 's--', label='Backtracking', linewidth=2)
plt.plot(df['n'], df['NumTransactions_DP'], '^:', label='DP Bitmask', linewidth=2)

plt.ylabel('Time (microseconds)')
plt.yscale('log')
plt.xlabel('Number of People (n)')
plt.ylabel('Number of Transactions')
plt.title('Transactions vs. Group Size')
plt.legend()
plt.grid(True)
plt.yscale('log')
plt.show()
```

```python


plt.plot(df['n'], df['Time_Greedy_us'], 'o-', label='Greedy')
plt.plot(df['n'], df['Time_Backtracking_us'], 's--', label='Backtracking')
plt.plot(df['n'], df['Time_DP_us'], '^:', label='DP Bitmask')
plt.xlabel('Number of People (n)')
plt.ylabel('Time (microseconds)')
plt.title('Algorithm Execution Time vs. Group Size')
plt.legend()
plt.grid(True)
plt.yscale('log')
plt.show()
```
